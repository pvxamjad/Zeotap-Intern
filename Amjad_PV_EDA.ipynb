```python
import pandas as pd
import matplotlib.pyplot as plt

customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')

# print(customers.isnull().sum())
# print(products.isnull().sum())
# print(transactions.isnull().sum())

# by this i found no missing data

# checked if there any duplications .
# print(customers.duplicated().sum())
# print(products.duplicated().sum()) 
# print(transactions.duplicated().sum())
# checked duplcation 
# that also zero 


# knowing count,unique,mean,top etc 
# print(customers.describe())
# print(products.describe())
# print(transactions.describe())

# checked the datatypes 
# print(customers.dtypes)
# print(products.dtypes)
# print(transactions.dtypes)


# Customers Analysis
# distribution by region 
# Visualizations with titles.
customers['Region'].value_counts().plot(kind='bar', title='Customers by Region')
plt.show()

# sign up trend over time 
# Visualizations with titles and labels.
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
customers['SignupDate'].dt.year.value_counts().sort_index().plot(kind='line', title='Customer Sign-Up Trends')

plt.xlabel('Year')  
plt.ylabel('Number of Sign-Ups')  
plt.xticks(rotation=45)  
plt.show()

 # Products Analysis

# most popular product catgories 
# Visualizations with titles.
products['Category'].value_counts().plot(kind='bar', title='Product Categories')
plt.show()

# price distribution
# Visualizations with titles.
products['Price'].plot(kind='line', title='Price Distribution')
plt.show()

# transaction analysis 
# top customer by toatal spend
top_customers = transactions.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)
print(top_customers)

# sales trend overtime 

#Data cleaning  coverting TransactionDate to Datetime
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'], errors='coerce')
# checking is it converted to datetime64[ns]
# print(transactions['TransactionDate'].dtype)
# also checking NaT
# print(transactions['TransactionDate'].isna().sum())

# Visualizations with titles and labels.
transactions.groupby(transactions['TransactionDate'].dt.date)['TotalValue'].sum().plot(
    kind='line', title='Sales Trends Over Time'
)
plt.xlabel('Date')
plt.ylabel('Total Sales Value')
plt.show()

```
